```python
from google.colab.output import eval_js
from IPython.display import display, Javascript
from PIL import Image
import io
import base64

def capture_image():
    js = Javascript('''
        async function captureImage() {
            const video = document.createElement('video');
            const div = document.createElement('div');
            document.body.appendChild(div);
            div.appendChild(video);

            const stream = await navigator.mediaDevices.getUserMedia({video: true});
            video.srcObject = stream;

            await new Promise((resolve) => video.onloadedmetadata = resolve);

            const canvas = document.createElement('canvas');
            canvas.width = video.videoWidth;
            canvas.height = video.videoHeight;
            const context = canvas.getContext('2d');
            context.drawImage(video, 0, 0, canvas.width, canvas.height);

            stream.getTracks().forEach(track => track.stop());
            div.remove();
            return canvas.toDataURL('image/jpeg');
        }
        ''')
    display(js)
    data = eval_js('captureImage()')
    binary = base64.b64decode(data.split(',')[1])
    return Image.open(io.BytesIO(binary))
```

```python
import streamlit as st
import cv2
import numpy as np
import pandas as pd
import pytesseract
from datetime import datetime
from ultralytics import YOLO
```

```python
st.title("Live License Plate Recognition App")
```

```python
model = YOLO("yolov8n.pt")
```

```python
run = st.checkbox('Start Camera')
FRAME_WINDOW = st.image([])
cap = cv2.VideoCapture(0)
```

```python
if 'plate_data' not in st.session_state:
    st.session_state.plate_data = pd.DataFrame(columns=['Plate Number', 'Time Detected'])
```

```python
while run:
    ret, frame = cap.read()
    if not ret:
        st.write("Camera not available")
        break

    results = model(frame)
    boxes = results[0].boxes

    for box in boxes:
        x1, y1, x2, y2 = map(int, box.xyxy[0])
        conf = float(box.conf[0])
        if conf < 0.5:
            continue

        plate_img = frame[y1:y2, x1:x2]
        gray = cv2.cvtColor(plate_img, cv2.COLOR_BGR2GRAY)
        thresh = cv2.threshold(gray, 100, 255, cv2.THRESH_BINARY)[1]
        plate_number = pytesseract.image_to_string(thresh, config='--psm 8').strip()

        if plate_number:
            timestamp = datetime.now().strftime("%Y-%m-%d %H:%M:%S")
            new_data = pd.DataFrame([{'Plate Number': plate_number, 'Time Detected': timestamp}])
            st.session_state.plate_data = pd.concat([st.session_state.plate_data, new_data], ignore_index=True)


        # Draw rectangle and put text on the frame
        cv2.rectangle(frame, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(frame, plate_number, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 255), 2)

    # Display the processed frame
    frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    FRAME_WINDOW.image(frame)
```

```python
if st.button("Show Detected Plates"):
    st.dataframe(st.session_state.plate_data)
```

```python
if st.button("Download CSV"):
    csv = st.session_state.plate_data.to_csv(index=False).encode('utf-8')
    st.download_button("Download Plate Records", data=csv, file_name="detected_plates.csv", mime="text/csv")
```
